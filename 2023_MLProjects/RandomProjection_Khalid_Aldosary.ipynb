## Executive Summary

   The problem is when having high dimensional dataset which will make it very expensive to perform calculation, or when having a real-time data, where accessibility to full data is insufficient. To address the problem, the method of random projection has been used to transform the data to lower dimensional space. What was learned is that points in a high-dimensionality space can be linearly embedded in a space of much lower dimensionality in such a way that distances between the points are nearly preserved. Recommendation is what Johnson-Lindenstrauss lemma result is, to use random projection when the number of data is small, and the dimensionality of the problem is very large.

## Full Code followed by result plot (please note that this is a repitition of the above codes, but I have summed it up all at once for convenience):







````python
import numpy as np                                        # for working with data and model arrays
import matplotlib.pyplot as plt                           # for plotting


# Generate gaussian reduction data set 

def gs_rp(data, num_p):
    rows, cols = data.shape
    rm = np.random.randn(cols, num_p)/np.sqrt(num_p)
    proj = np.dot(data,rm)
    return proj,rm

# Compute projection error

def proj_err(orig, proj, rand):
    recon = np.dot(proj, np.linalg.pinv(rand))
    err = np.linalg.norm(orig-recon)/np.linalg.norm(orig)
    return err

# Generate Random Data of 100 Rows and 50 Columns

data = np.random.rand(100,50)
num_proj = np.arange(1, data.shape[1]+1)
errs = []

# Loop over the number of columns to perfrom gassuain reduction and append to errors vector

for prj in num_proj:
    prj_data, rand_m = gs_rp(data, prj)
    error = proj_err(data, prj_data, rand_m)
    errs.append(error)
    
#Plotting the results

plt.plot(num_proj, errs, marker='o')
plt.title('Projection Error vs Number of Projections')
plt.xlabel('Number of Projections (p)')
plt.ylabel('Projection Error')
plt.grid(True)
plt.axvline(x = 50, color = 'red') # Indiacting the original dimension on the results plot
plt.text(50.6,0,'Original Dimension',rotation=90)
plt.show()
````

### Discussing the Results:

As can be seen from the plot, as the number of projection (p) decreases (by utilizing gaussian random projection), the projection error increases.Noting that the number of projection starts from 50 with 0 projection error as 50 is the original dimension. It is now up to user to determine what projection error is acceptable and use the number of projection opposite to it. What was learned is that points in a high-dimensionality space can be linearly embedded in a space of much lower dimensionality in such a way that distances between the points are nearly preserved. Recommendation is what Johnson-Lindenstrauss lemma result is, to use random projection when the number of data is small, and the dimensionality of the problem is very large.
