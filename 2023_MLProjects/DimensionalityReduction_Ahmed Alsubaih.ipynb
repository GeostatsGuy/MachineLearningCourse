### Executive Summary
The dimensional reduction is one of the most effective techniques to deal with huge data sets to improve the processes of data manipulation. It can be effective to reduce the computational time and mitigate the redundant in the data. Basically, there is two techniques being implemented for dimensional reduction that is feature selection and dimensional reduction. In the first approach (feature selection) is merely keeping only the most relevant features in the data set while in the dimensional reduction is finding the feature that has similar information that has in the input data. The project isto invistigate  the dimensional reduction methods. There are several techniques that use in the dimensional reduction of the data. The primary target is to ensure accuracy in the input data and reduce the computational time.  I will compare the prediction and the variances that I got for each technique by using the data from the class and compare it with class results. ## 1-Mutual Information:

Methodology: Mutual information is a metric used to quantify the amount of information shared between two variables. In the context of feature selection, it measures the dependency between individual features and the target variable.

Interpretation: Mutual information calculates the reduction in uncertainty about the target variable based on the knowledge of a feature's value. Higher mutual information values indicate a stronger relationship between the feature and the target. Features with high mutual information are considered more informative for predicting the target variable.

## 2-Random Forest Feature Importance:

Methodology: RandomForestRegressor is used to fit the training data and calculate the feature importances.
Interpretation: Feature importances represent the contribution of each feature in predicting the target variable. Higher importances indicate more influential features.

## 3-Low Variance Filter:

Methodology: VarianceThreshold is applied to filter out features with variance below a specified threshold.
Interpretation: Features with low variance are considered less informative, and this method helps eliminate them.

## 4-Principal Component Analysis (PCA):

Methodology: PCA is a linear dimensionality reduction technique that transforms the data into a new set of uncorrelated variables (principal components).
Interpretation: Principal components capture the maximum variance in the data. The first few components often represent the most important patterns.

## 5-Independent Component Analysis (ICA):

Methodology: FastICA is used to perform ICA, a technique for separating a multivariate signal into additive, independent components.
Interpretation: Independent components aim to represent statistically independent sources in the data, uncovering hidden factors.
## 6-t-Distributed Stochastic Neighbor Embedding (t-SNE):

Methodology: t-SNE is a non-linear dimensionality reduction technique that emphasizes preserving local structure in the data.
Interpretation: t-SNE is effective for visualizing clusters in the data. Points close in the reduced space reflect similar local relationships. The color of each point in the t-SNE plot is determined by the values of the selected feature . Points with similar colors have similar values for the selected feature.

## 7-Uniform Manifold Approximation and Projection (UMAP):

Methodology: UMAP is a non-linear dimensionality reduction technique that also focuses on preserving local and global structure.
Interpretation: UMAP is particularly effective for preserving both local and global structures, often providing better representations of data clusters.Similar to t-SNE, the color of each point in the UMAP plot represents the values of the selected feature. Points with similar colors share similar feature values.

## Explained Variance:

Methodology: The explained variance for PCA and ICA is calculated based on the variance captured by the components. For t-SNE and UMAP, it's set to 1.0 as these methods don't provide explicit explained variance.
Interpretation: Explained variance represents the proportion of the total variance in the data that is captured by the reduced dimensions. Higher values indicate more information retained

## 2- Workflow to compare ICA and PCA methods
This code performs Principal Component Analysis (PCA) and Independent Component Analysis (ICA) on a dataset with the specified features. It then creates an interactive widget for selecting different features for the x and y axes. The code also includes the restoration of data by setting the second component to zero and then reversing the transformation.

Let's break down the results and the functionality:

Feature Selection and Standardization:

The code starts by selecting specific features ('Por', 'Perm', 'AI', 'Brittle', 'TOC', 'VR') and standardizing the data.
PCA and ICA:

PCA is applied to reduce the dimensionality to two components.
ICA is applied to extract independent components.
Interactive Widget:

An interactive widget is created using the interact function, allowing users to select features for the x and y axes.
Two subplots are displayed: one for PCA and one for ICA.
The color of the points in the scatter plots is determined by the selected feature for the x-axis.
Restoration of Data (PCA):

The second principal component is removed by setting its scores to 0.0.
The reversed principal component is calculated by matrix multiplication and restoration is performed.
The restored data is visualized in a scatter plot, where the color represents the selected feature for the x-axis.
Restoration of Data (ICA):

Similar to PCA, the second independent component is removed by setting its scores to 0.0.
The reversed independent component is calculated by matrix multiplication and restoration is performed.
The restored data is visualized in another scatter plot, with color representing the selected feature for the x-axis.
In summary, the code allows users to interactively explore how different features influence the scatter plots generated by PCA and ICA. The restoration of data provides insight into the impact of removing specific components on the overall dataset. Users can select different features for the x-axis in both the interactive widget and the restoration plots, facilitating a dynamic exploration of the data.
## The expain varaince for each method also presented as well as the original mean for the selected features.

```python
# Select the specific features for PCA and ICA
all_features = data.columns.tolist()
selected_features = ['Por', 'Perm', 'AI', 'Brittle', 'TOC', 'VR']

# Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(data[selected_features])

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_std)

# ICA
ica = FastICA(n_components=2)
X_ica = ica.fit_transform(X_std)

# Create an interactive widget for selecting x and y axes
@interact(x_axis=all_features, y_axis=all_features)
def plot_dimensionality_reduction(x_axis, y_axis):
    # Create subplots for PCA and ICA
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # Plot PCA
    scatter_pca = axes[0].scatter(X_pca[:, 0], X_pca[:, 1], c=data[x_axis], cmap='viridis')
    axes[0].set_title('PCA - Principal Component Scores')
    axes[0].set_xlabel('Principal Component 1')
    axes[0].set_ylabel('Principal Component 2')
    plt.colorbar(scatter_pca, ax=axes[0], label=x_axis)

    # Plot ICA
    scatter_ica = axes[1].scatter(X_ica[:, 0], X_ica[:, 1], c=data[x_axis], cmap='viridis')
    axes[1].set_title('ICA - Independent Component Scores')
    axes[1].set_xlabel('Independent Component 1')
    axes[1].set_ylabel('Independent Component 2')
    plt.colorbar(scatter_ica, ax=axes[1], label=x_axis)

    plt.tight_layout()
    plt.show()

    # Remove the 2nd principal component by setting associated component scores to 0.0
    X_pca_no_pc2 = X_pca.copy()
    X_pca_no_pc2[:, 1] = 0.0

    # Reverse the principal component by matrix multiplication
    X_pca_reversed = np.dot(X_pca_no_pc2, pca.components_) + pca.mean_

    # Print Explained Variance for PCA
    explained_variance_pca = np.sum(pca.explained_variance_ratio_)
    print(f'Explained Variance (PCA) for {x_axis}: {explained_variance_pca:.4f}')

    # Print Original Mean for PCA
    original_mean_pca = scaler.inverse_transform([pca.mean_])
    print(f'Original Mean (PCA) for {x_axis}: {original_mean_pca[0][all_features.index(x_axis)]:.4f}')

    # Visualize the restored data
    plt.figure(figsize=(10, 5))
    scatter_restored = plt.scatter(X_pca_reversed[:, 0], X_pca_reversed[:, 1], c=data[x_axis], cmap='viridis')
    plt.title('Restored Data PCA')
    plt.xlabel(f'Restored {x_axis}')
    plt.ylabel(f'Restored {y_axis}')
    plt.colorbar(scatter_restored, label=x_axis)
    plt.show()

    # Remove the 2nd independent component by setting associated component scores to 0.0
    X_ica_no_ic2 = X_ica.copy()
    X_ica_no_ic2[:, 1] = 0.0

    # Reverse the independent component by matrix multiplication
    X_ica_reversed = np.dot(X_ica_no_ic2, ica.mixing_.T) + ica.mean_

    # Print Explained Variance for ICA
    explained_variance_ica = np.sum(np.var(X_ica_reversed, axis=0)) / np.sum(np.var(X_std, axis=0))
    print(f'Explained Variance (ICA) for {x_axis}: {explained_variance_ica:.4f}')

    # Print Original Mean for ICA
    original_mean_ica = scaler.inverse_transform([ica.mean_])
    print(f'Original Mean (ICA) for {x_axis}: {original_mean_ica[0][all_features.index(x_axis)]:.4f}')

    # Visualize the restored data
    plt.figure(figsize=(10, 5))
    scatter_restored_ica = plt.scatter(X_ica_reversed[:, 0], X_ica_reversed[:, 1], c=data[x_axis], cmap='viridis')
    plt.title('Restored Data (ICA)')
    plt.xlabel(f'Restored {x_axis}')
    plt.ylabel(f'Restored {y_axis}')
    plt.colorbar(scatter_restored_ica, label=x_axis)
    plt.show()
```
